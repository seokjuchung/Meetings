```python
import pandas as pd

import matplotlib.pyplot as plt

# Parse the conference data
conferences = [
    {'year': 2024, 'name': 'New Perspectives', 'type': 'domestic', 'topic': 'RAD'},
    {'year': 2024, 'name': 'CPAD', 'type': 'domestic', 'topic': 'RAD'},
    {'year': 2025, 'name': 'APS', 'type': 'domestic', 'topic': 'SBND'},
    {'year': 2025, 'name': 'AD4HEP', 'type': 'domestic', 'topic': 'RAD'},
    {'year': 2025, 'name': 'Users Meeting', 'type': 'domestic', 'topic': 'RAD'},
    {'year': 2025, 'name': 'NPML', 'type': 'international', 'topic': 'RAD'},
    {'year': 2025, 'name': 'CTD', 'type': 'international', 'topic': 'RAD'},
]

df_conf = pd.DataFrame(conferences)

# Create figure with two subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram 1: Domestic vs International
conf_type_counts = df_conf['type'].value_counts()
conf_type_counts.plot(kind='bar', ax=axes[0], color=['steelblue', 'coral'])
axes[0].set_title('Number of Domestic vs International Conferences')
axes[0].set_ylabel('Count')
axes[0].set_xlabel('Conference Type')
axes[0].tick_params(axis='x', rotation=0)

# Histogram 2: By Topic
topic_counts = df_conf['topic'].value_counts()
topic_counts.plot(kind='bar', ax=axes[1], color=['steelblue', 'coral'])
axes[1].set_title('Number of Conferences by Topic')
axes[1].set_ylabel('Count')
axes[1].set_xlabel('Topic')
axes[1].tick_params(axis='x', rotation=0)

# Histogram 3: By Year
year_counts = df_conf['year'].value_counts().sort_index()
year_counts.plot(kind='bar', ax=axes[2], color=['steelblue', 'coral'])
axes[2].set_title('Number of Conferences by Year')
axes[2].set_ylabel('Count')
axes[2].set_xlabel('Year')
axes[2].tick_params(axis='x', rotation=0)

plt.savefig('plots/conference_plot.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
